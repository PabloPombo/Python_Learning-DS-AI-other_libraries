## Box plot

```python
np.random.seed(10)
data_1 = np.random.normal(100,10,200)
data_2 = np.random.normal(80,30,200)
data_3 = np.random.normal(70,20,200) # numeros al azar
```

```python
data_to_plot = [data_1, data_2, data_3]
np.shape(data_to_plot)
```

```python
np.size(data_to_plot)
```

```python
bp = plt.boxplot(data_to_plot)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,9))
#boxplot rectangular
bplot1 = axes[0].boxplot(data_to_plot, vert=True, patch_artist=True)
bplot2 = axes[1].boxplot(data_to_plot, notch=True, vert=True, patch_artist=True)
              
#Colores de la paleta
colors=['pink', 'lightblue', 'lightgreen']
for bplot in (bplot1,bplot2):
    for patch, color in zip(bplot['boxes'],colors):
        patch.set_facecolor(color)
    
#Adicionando nombres a abscisas 
plt.setp(axes, xticks=[y+1 for y in range(len(data_to_plot))], xticklabels= ['x1','x2','x3'])
```

```python
import requests as rq
import pandas as pd
import csv

Data = rq.get("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")

with open('iris.data','w') as f:
     f.write(Data.text)

with open('iris.data') as f:
    reader = csv.reader(f)
    iris_data = [x for x in reader]   
            
iris_DataFrame = pd.DataFrame(iris_data)
iris_DataFrame.columns = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width', 'Class']
iris_DataFrame
```

```python
iris_DataFrame.head()
```

```python
iris_DataFrame.sort_values(['Sepal_Length'], ascending=False) #acomodar 
```

```python
datos = pd.concat([iris_DataFrame['Sepal_Length'], iris_DataFrame['Sepal_Width']], axis=1)
```

```python
datos = pd.concat([iris_DataFrame['Sepal_Length'].astype('float'), iris_DataFrame['Sepal_Width'].astype('float'),
                   iris_DataFrame['Petal_Length'].astype('float'), iris_DataFrame['Petal_Width'].astype('float'),
                   iris_DataFrame['Class'].astype('str')], axis=1)
```

```python
datos = pd.DataFrame(datos)
```

```python
datos.dropna(axis=0)
```
