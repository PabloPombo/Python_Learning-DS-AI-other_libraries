```python
import pandas as pd
from pandas_datareader import data as wb
```

```python
tickers = ['AAPL', 'MSFT', 'XOM', 'BP']
new_data = pd.DataFrame()
for t in tickers:
    new_data[t] = wb.DataReader(t, data_source='yahoo', start='2015-1-1')['Close']
```

```python
new_data.to_csv('D:\ejemplo.csv')
```
