```python
import cv2
import pytesseract

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract'  #Read tesseract

path = r'D:\Fotos-Imagenes\imagen.jpg'   #Path Image

cv2.namedWindow('image', cv2.WINDOW_NORMAL)     #Create window with freedom of dimensions
cv2.resizeWindow('image', 1000, 800)            # Resize window to specified dimensions

image = cv2.imread(path)                        # Read Image

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  #Gray
canny = cv2.Canny(gray, 10, 150)                #Border detection
canny = cv2.dilate(canny, None, iterations = 1) #Engrosar bordes

cnts = cv2.findContours(canny, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[0]        #Find External Contourns
cnts = sorted(cnts, key=cv2.contourArea, reverse=True)[:]                            #Ordenar contornos 


for c in cnts:
    epsilon = 0.01*cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, epsilon, True)
    #print ('approx = ', approx)
    
    if len(approx) >= 4:
        cv2.drawContours(image, [approx], 0,(255,0,0),2)#[:1]
   

text = pytesseract.image_to_string(canny)       #Image to string

#if (text=='Pantalla'):
print('', text)                                 #Print Text

#window_name = 'image'
cv2.imshow('image', image)                      #Show Image
cv2.imshow('canny', canny)  

cv2.waitKey(0)                                  #Display image until Keypress
cv2.destroyAllWindows()
```

```python
import cv2
import pytesseract
from pytesseract import Output

myconfig = r'--psm 11 --oem 3'
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract'  #Read tesseract

#path = r'C:\Users\Pablo\Desktop\imagen6.jfif'   #Path Image
path = r'D:\Fotos-Imagenes\imagen.jpg' 

cv2.namedWindow('image', cv2.WINDOW_NORMAL)     #Create window with freedom of dimensions
cv2.resizeWindow('image', 1000, 800)            # Resize window to specified dimensions
#cv2.namedWindow('canny', cv2.WINDOW_NORMAL)     #Create window with freedom of dimensions
#cv2.resizeWindow('canny', 1000, 800)            # Resize window to specified dimensions

image = cv2.imread(path)                        # Read Image
height, width, _ = image.shape

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  #Gray
canny = cv2.Canny(gray, 10, 150)                #Border detection
canny = cv2.dilate(canny, None, iterations = 1) #Engrosar bordes
   
text = pytesseract.image_to_string(canny)       #Image to string
data = pytesseract.image_to_data(canny, config=myconfig, output_type=Output.DICT)

value = "Owner's"

amount_boxes = len(data['text']) 
for i in range(amount_boxes):
    if float (data['conf'][i]) > 50:
        if data['text'][i] == value:
            (x, y, width, height) = (data['left'][i], data['top'][i], data['width'][i], data['height'][i])   #Obtener valores de dato y lugar donde se encuentra
            image = cv2.rectangle(image, (x, y), (x+width, y+height), (0,255,0),2)                           #Colocar rectangulo
            image = cv2.circle(image, (x+int(width/2), y+int(height/2)), 5, (0,255,0),2)                     #Colocar circulo en el centro de rectangulo
            image = cv2.putText(image, data['text'][i], (x, y+height+20), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,0),2, cv2.LINE_AA)   #Colocar el Texto

print('', text)                                 #Print Text
print(data['text'])   

#window_name = 'image'
cv2.imshow('image', image)                      #Show Image
cv2.imshow('canny', canny)  

cv2.waitKey(0)                                  #Display image until Keypress
cv2.destroyAllWindows()
```
